```python
#Removing non-wanted variables by commenting them out

df = df[['coaster_name', 
#'Length', 'Speed', 
'Location', 'Status', 
#'Opening date', 'Type',
 'Manufacturer', 
 #'Height restriction', 'Model', 'Height',
  #     'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
   #    'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #   'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
     #  'Track layout', 'Fastrack available', 'Soft opening date.1',
      # 'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # '#Single rider line available', 'Restraint Style',
      # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
       # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
       #'height_value', 'height_unit', 
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()

# columns can also be dropped with df.drop([''])
```

```python
#opening date is an object, we have to change it to datetime

df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
#changing the year from integer  to numeric

df['year_introduced'] = pd.to_numeric(df['year_introduced']) 
```

```python
#Renaming some of the columns to proper names and re-writing the df name
df = df.rename(columns = {'coaster_name': 'Coaster_Name',
                    'year_introduced': 'Year_Introduced',
                    'opening_date_clean': 'Opening Date',
                    'speed_mph': 'Speed_mph',
                    'height_ft': 'Height_ft',
                    'Inversions_clean': 'Inversions',
                    'Gforce_clean': 'Gforce'})
```

```python
#To identify missiong values in each column of the dataframe

df.isna().sum()
```

```python
#To checkk if there is duplicated records(row) in the dataframe

#df.duplicated() #'Coaster_Name'
 
df.loc[df.duplicated(subset=['Coaster_Name'], keep=False)]

#df.loc[df.duplicated()].shape

#No duplication of any record in the dataframe
```

```python
#Even though there is no duplication of entire records, there could be dupplication of coaster names, running the duplicate for only 'Coaster_Name' to find if duplicated Coaster names exist

df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening Date'])].reset_index(drop=True).copy()
       

#print(df)

#97 rows have a duplicated coaster name on first run, our index got scattered due to the drop in rows, to maintain that, we do a .reset_index() method 
```

## 3. Feature Understanding
<p> (Univariate analysis): One dimensional/variable charts and understanding

- Plotting Feature Distributions
    - Histogram
    - KDE (Density Plot)
    - Boxplot


```python
#imported seaborn package to be used
#Visualizing the year that the roller coasters where build using a barchart. Top 20 years are visualized

ax = df['Year_Introduced'].value_counts().head(20).plot(kind = 'bar', title = 'Top 20 Years that Coaster was introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
plt.show()
```

```python
#Visualizing the Speed distribution of the rollercoasters with a histogram

ax = df['Speed_mph'].plot(kind = 'hist', bins = 20, title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show()
```

```python
#A kde plot of the same speed mph instead

ax = df['Speed_mph'].plot(kind = 'kde',  title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show()
```

```python
# To look at correlations between certain features (only on numeric variables), NaN values are dropped because matrix plots such as the heatmap require that the data is 'tidy'

df_corr =df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
df_corr
```

```python
#Visualizing the heatmap;
sns.heatmap(df_corr, annot= True) 
```

## 5. Answering a Question on the Dataframe

<li> What are the locations with the fastest roller coasters (minimum of ten coasters on that location)?

```python
#'Others' appear as a list of location, we are going to exclude this and group data based on location
ax = df.query('Location != "Other" ')\
        .groupby('Location')['Speed_mph'] \
        .agg(['mean', 'count']) \
        .query('count >= 10') \
        .sort_values('mean', ascending=True) \
        .plot(kind = 'barh', figsize=(12,10), title = 'Average Coastal Speed and Count by Location')

ax.set_xlabel('Averae Coaster Speed and Count')
plt.show()
```
